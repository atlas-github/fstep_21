#Build loop for TSLA, GOOGL, TWTR, and MSFT

```python
import requests

url = "http://api.marketstack.com/v1/eod?access_key=d59eea61a0e67c2d6990a49b17e37857&symbols=TWTR"

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)

c_json = json.loads(response.text)
df3 = pd.json_normalize(c_json["data"])
df3
```

```python
df_compiled = pd.DataFrame(columns = ["open", "high", "low", "close", "volume", "adj_high", "adj_low", "adj_close", "adj_open", "adj_volume", "split_factor", "dividend", "symbol", "exchange", "date"])
df_compiled
```

```python
stocks = ["TSLA", "GOOGL", "TWTR", "MSFT"]

for i in range(0, len(stocks)):
  print(stocks[i])
```

```python
for i in range(0, len(stocks)):
  url = "http://api.marketstack.com/v1/eod?access_key=d59eea61a0e67c2d6990a49b17e37857&symbols="+stocks[i]

  payload={}
  headers = {}

  response = requests.request("GET", url, headers=headers, data=payload)

  c_json = json.loads(response.text)
  df3 = pd.json_normalize(c_json["data"])
  df_compiled = df_compiled.append(df3)
```
